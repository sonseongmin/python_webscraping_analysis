```python
import requests
import json
import pandas as pd
import os
# 네이버 API 인증 정보
client_id = "x_45StwzjCiPnscCfTqJ"
client_secret = "aOhY7CBjXw"

# 공통 헤더
headers = {
    'X-Naver-Client-Id': client_id,
    'X-Naver-Client-Secret': client_secret
}
os.makedirs('data',exist_ok=True)
# 검색 함수 정의
def search_shops(query):
    url = "https://openapi.naver.com/v1/search/shop.json"
    
    # 쿼리 파라미터 설정
    payload = {
        'query': query,
        'display': 10,
        'sort': 'sim'
    }

    # API 요청
    res = requests.get(url, params=payload, headers=headers)

    # 응답 코드 확인
    if res.status_code == 200:
        items = res.json()['items']
        return items
    else:
        print("Error:", res.status_code)
        return []


# 결과 파싱 예시
shops = search_shops("가디건")
with open('data/shops.json', 'w', encoding='utf-8') as f:
    json.dump(shops, f, ensure_ascii=False, indent=2)
df= pd.read_json('data/shops.json')
df
# 파싱해서 출력
#for book in books:
 #   print("제목:", book['title'])
 #   print("저자:", book['author'])
  #  print("출판사:", book['publisher'])
   # print("링크:", book['link'])
    #print("-" * 40)
#df = pd.read_json('data/books.json')
#df
```
